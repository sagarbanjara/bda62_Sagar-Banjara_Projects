```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans     #importing necessary libraries for visualization
```

```python
plt.scatter(df['Age'], df['Spending Score (1-100)'])
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('mall customer')        #giving the specific name for x axis and y axis and title.
plt.show()
```

```python
plt.scatter(df['Age'], df['Annual Income (k$)'])
plt.xlabel('Annual Income')
plt.ylabel('Age')
plt.show()
```

```python
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters =  k)
    kmeans.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']])
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10), inertia)     #indicating the range 
plt.xlabel('Number of clusters')
plt.ylabel("inertia")
```

```python
kmeans = KMeans(n_clusters = 4)
predicted = kmeans.fit_predict(df[['Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
predicted
```

```python
df['Cluster'] = predicted     #checking the number of cluster
```

```python
df
```

```python
df1 = df[df.Cluster==0]    #defining the df for each cluster
df2 = df[df.Cluster==1] 
df3 = df[df.Cluster==2]
df4 = df[df.Cluster==3]
```

```python
plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],color='grey')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],color='green')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],color='blue')
plt.scatter(df4['Annual Income (k$)'],df4['Spending Score (1-100)'],color='orange')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='red',marker='*',label='centroid')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.title('mall customer')
plt.show()
```
